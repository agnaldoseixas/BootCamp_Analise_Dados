```python
import mysql.connector
import pandas as pd
```

```python
#configurações
host = '34.70.39.79' #Endereço do banco que queremos acessar
user = 'root' # Usuário padrão
password = '' # Se for sem senha, é só deixar assim
database = 'faculdade'
```

```python
!curl ipecho.net/plain # retorna o ip do colab
```

```python
'''
Criando a conexão entre o nosso código e um banco mysql
Nós chamamos essa função: mysql.connector.connect() e passamos os parâmetros
necessários para realizar a conexão.
host: Endereço do servidor
user: Nome do usuário no banco
password: Senha do usuário
database: Nome do banco
a variável connection guarda o resultado da conexão
'''
connection = mysql.connector.connect(
    host = host,
    user = user,
    password = password,
    database = database
)
cursor = connection.cursor() # O cursor fica apontando para o banco de dados.
query = 'SHOW TABLES;' # A variável query armazena a consulta desejada.
cursor.execute(query) # Executa a query
result = cursor.fetchall() # Armazenando o resultado da consulta.
tabelas = pd.DataFrame(result,columns=cursor.column_names) # convertendo o result num dataframe pandas
display(tabelas) # exibindo o dataframe
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
connection = mysql.connector.connect(
    host = host,
    user = user,
    password = password,
    database = database
)
cursor = connection.cursor() # O cursor fica apontando para o banco de dados.
query = '''
SELECT * FROM Aluno;
''' # A variável query armazena a consulta desejada.
cursor.execute(query) # Executa a query
result = cursor.fetchall() # Armazenando o resultado da consulta.
aluno = pd.DataFrame(result,columns=cursor.column_names) # convertendo o result num dataframe pandas
display(aluno) # exibindo o dataframe
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```
