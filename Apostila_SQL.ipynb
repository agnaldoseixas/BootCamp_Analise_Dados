```python
def executar_consulta(query):
  try:
    connection = mysql.connector.connect(
        host=host,
        user=user,
        password=password,
        database=database
    ) # Criando a conexão com o banco
    cursor = connection.cursor() # Criando um cursor que aponta para o banco
    cursor.execute(query) # Executando a query
    result = cursor.fetchall() # Armazenando os resultados da minha consulta
    tabelas = pd.DataFrame(result,columns = cursor.column_names) # Convertendo o resultado para um dataframe
    return tabelas

  except mysql.connector.Error as err:
    print(f'Erro: {err}')
    return None

  finally:
    if cursor:
      cursor.close()
    if connection:
      connection.close()
```

```python
query = 'SHOW TABLES;'
resultado = executar_consulta(query)
display(resultado)
```

#Booleanos (E | OU | NÃO)

```python
'''
AND - Todas as condições devem ser verdadeiras

Exemplo: Retorne todos os clinetes do sexo masculino da cidade de Curitiba
'''
query='''
SELECT nome, cidade, genero
FROM clientes
WHERE genero = 'Masculino' AND cidade = 'Curitiba';
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
'''
OR - Pelo menos uma das condições precisa ser verdadeira.

Exemplo: Retorne todos os clientes que moram em Curitiba OU em São Paulo
'''
query='''
SELECT nome, cidade, genero
FROM clientes
WHERE cidade = 'Curitiba' OR cidade = 'São Paulo';
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
'''
NOT - Negação

Exemplo: Queremos todos os clientes que não são de São Paulo
'''
query='''
SELECT nome, cidade, genero
FROM clientes
WHERE NOT cidade = 'São Paulo';
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
'''
Retorne todos os clientes que não sejam de Curitiba e não sejam de São Paulo
'''
query='''
SELECT nome, cidade, genero
FROM clientes
WHERE NOT cidade = 'São Paulo' AND NOT cidade = 'Curitiba';
'''
resultado = executar_consulta(query)
display(resultado)
```

#9 - Group By

```python
'''
No group by nós agrupamos linhas com base nos valores de uma ou mais colunas
(dimensões/campos cetegoricos). Com isso, podemos realizar operações de agregação
(sum, avg, max, min, count)

Exemplo: Média do salário agrupada por gênero
'''
query='''
SELECT genero, AVG(salario) AS "Média Salarial"
FROM funcionarios
GROUP BY genero;
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
query='''
SELECT estado_civil, AVG(salario) AS "Média Salarial"
FROM funcionarios
GROUP BY estado_civil;
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
query='''
SELECT genero, estado_civil, ROUND(AVG(salario),2) AS "Média Salarial"
FROM funcionarios
GROUP BY genero, estado_civil;
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
query='''
SELECT genero, cargo, COUNT(id_funcionario) AS "Qtd de Funcionários"
FROM funcionarios
GROUP BY genero, cargo;
'''
resultado = executar_consulta(query)
display(resultado)
```

#10 - HAVING

```python
'''
Usamos o Having para filtrar um resultado após a aplicação de funções de agregação.
Normalmente é utilizado junto com o Group By, ele tem como função limitar
os grupos retornados com base em alguma condição.

WHERE: Filtra antes do agrupamento. O Where não permite o uso de funções de
agregação dentro dele.

Having: Filtra depois do agrupamento. Permite usar funções de agregação dentro dele.
'''
query='''
SELECT nome, salario
FROM funcionarios
HAVING salario > 3500;
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
query='''
SELECT genero, MIN(salario) AS "Menor Salário"
FROM funcionarios
GROUP BY genero
HAVING MIN(salario) > 2800;
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
# Não conseguimos realizar a mesma operação com o Where
query='''
SELECT genero, MIN(salario) AS "Menor Salário"
FROM funcionarios
WHERE MIN(salario)>2800
GROUP BY genero;
'''
resultado = executar_consulta(query)
display(resultado)
```

#12 - IN | BETWEEN

```python
'''
IN: Usamos para verificar se um valor esta presente em uma lista de valores
especificadas. É mais prático do que utilizar múltiplos operadores OR.

Se quisessemos selecionar os produtos com id 2, 4 e 6.

Exemplo sem o IN
'''
query='''
SELECT nome, id_produto
FROM produtos
WHERE id_produto = 2 OR id_produto = 4 OR id_produto = 6
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
'''
Exemplo com o IN
'''
query='''
SELECT nome, ID_produto
FROM produtos
WHERE ID_produto IN (2,4,6);
'''
resultado = executar_consulta(query)
display(resultado)
```

```python
'''
BETWEEN: Usamos para filtrar os resultados com base em um intervalo de valores,
podemos trabalhar com números, com datas, o importe é que seja ordenável. Os
limites são inclusivos, então o começo e o fim do intervalo também faz parte da
saída.

Exemplo: Queremos trazer os produtos cujo ID vão de 1 até 9
'''
query='''
SELECT nome, id_produto
FROM produtos
WHERE id_produto BETWEEN 1 AND 9;
'''
resultado = executar_consulta(query)
display(resultado)
```

#17 - REPLACE

```python
'''
REPLACE: Usamos para substituir um caracter ou um conjunto de caracteres.

REPLACE(campo, 'caracter que vai ser substituido', 'o que vai entrar no lugar do caracter')
'''
query='''
SELECT cargo, REPLACE(cargo, 'Caixa', 'Vendedor') AS "Cargo Atualizado"
FROM funcionarios
ORDER BY cargo ASC;
'''
resultado = executar_consulta(query)
display(resultado)
```

#23 - CASE WHEN

```python
'''
O case when é basicamente uma estrutura condicional dentro de uma consulta SQL,
normalmente é utilizado para gerar uma nova coluna ou para categorizar os dados.

CASE
  WHEN condição1 THEN alguma_coisa
  WHEN condição2 THEN outra_coisa
  ...
  ELSE valor_padrao
END AS 'Apelido qualquer'
'''
query='''
SELECT id_venda, quantidade,
CASE WHEN quantidade > 2 THEN 'Atacado'
ELSE 'Varejo' END AS 'Classificação'
FROM vendas;
'''
resultado = executar_consulta(query)
display(resultado)
```

#24 - VIEW

```python
'''
VIEW: Consulta armazenada no banco de dados, também é chamada de tabela virtual
ou de visão. Usamos para economizar tempo com consultas que são feitas
frequentemente.

CREATE OR REPLACE VIEW exemplo AS
SELECT nãnaãnã
FROM nañãnã
WHERE nãnãnã;
'''
query='''
CREATE OR REPLACE VIEW exemplo AS
SELECT *
FROM vendas;
'''
resultado = executar_consulta(query)
display(resultado)
```
