```python
import mysql.connector
import pandas as pd
```

```python
#configurações
host = '35.202.62.190' #Endereço do banco que queremos acessar
user = 'root' # Usuário padrão
password = '' # Se for sem senha, é só deixar assim
database = 'escola'
```

```python
connection = mysql.connector.connect(
    host = host,
    user = user,
    password = password,
    database = database
)
cursor = connection.cursor() # O cursor fica apontando para o banco de dados.
query = '''
SELECT * FROM alunos;
''' # A variável query armazena a consulta desejada.
cursor.execute(query) # Executa a query
result = cursor.fetchall() # Armazenando o resultado da consulta.
alunos = pd.DataFrame(result,columns=cursor.column_names) # convertendo o result num dataframe pandas
display(alunos) # exibindo o dataframe
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```
