```python
import pandas as pd

pd.set_option('display.max_rows',1000)
```

```python
df = pd.read_csv('/content/Preços semestrais - AUTOMOTIVOS_2024.02.csv',
                 encoding='latin1',
                 sep = ';')
display(df)
```

```python
excluir = ['CNPJ da Revenda',
           'Nome da Rua',
           'Numero Rua',
           'Complemento',
           'Bairro',
           'Cep',
           'Valor de Compra']
# df.drop(dados a serem excluidos, eixo) axis = 1 a operação é na coluna
df = df.drop(excluir, axis=1)
```

```python
'''
dropna: Remove valores ausentes (NaN)
how = all: Só remove se todos os valores forem NaN
'''
# df.dropna(how='all', inplace=True)
df = df.dropna(how='all')
display(df)
```

```python
df.rename(columns={'ï»¿Regiao - Sigla':'Regiao','Estado - Sigla':'Estado'},inplace=True)
df.head()
```

```python
import numpy as np

def removeVirgula(num):
  try:
    num = str(num).replace(',','.') # transformando pra uma string e trocando a , por um .
    return float(num) # Transformando para um float
  except:
    # Sempre que tivermos um valor inválido ('abcfhk') retorna um valor nulo (np.NaN)
    return np.NaN

#A função map() simplifica a aplicação de uma função a uma sequência de elementos, evitando a necessidade de loops explícitos como FOR
df['Valor de Venda'] = df['Valor de Venda'].map(removeVirgula)

# df['Valor de Venda'] = pd.to_numeric(df['Valor de Venda'].astype(str).str.replace(',','.'), errors = 'coerce')
```

```python
# Convertendo a data para datetime
# errors = 'coerce' força a transformação se o valor for invalido, retorna NaT
# NaT significa Not a Time
df['Data da Coleta'] = pd.to_datetime(df['Data da Coleta'], format='%d/%m/%Y', errors='coerce')
display(df)
```

# Qual estado tem o registro de gasolina mais cara?

# Descubra quais estados tem o menor preço medio do etanol

```python
'''
Retornando a média, o valor mínimo e o valor máximo do etanol de cada estado.
Estamos organizando da menor média do valor de venda para a maior. (TOP 10)
'''
df_etanol = df[df['Produto'] == 'ETANOL']
df_etanol.groupby(['Estado'])['Valor de Venda'].mean().sort_values(ascending=True).head(10)
```

```python
df_etanol.groupby(['Estado'])['Valor de Venda'].agg(['mean','min','max']).sort_values(by ='mean', ascending=True).head(10)
```

```python
resumo_etanol = df_etanol.groupby(['Estado'])['Valor de Venda'].agg(['mean','min','max']).sort_values(by ='mean', ascending=True).head(10)
display(resumo_etanol)
```

```python
resumo_etanol['variacao'] = resumo_etanol['max'] - resumo_etanol['min']
display(resumo_etanol)
```
