# Medidas de Tendencia Central - Inativos

```python
# Pegando as medidas de idade
media = df_inativos['Idade'].mean() #media
mediana = df_inativos['Idade'].median() #mediana
# A moda retorna uma série, por isso utilizamos o [0] para retornar o primeiro valor.
moda = df_inativos['Idade'].mode()[0] #moda

print(f'Média: {media}')
print(f'Mediana: {mediana}')
print(f'Moda: {moda}')
```

# Medidas de Dispersão - Inativos

```python
variancia = df_inativos['Idade'].var() #variancia da idade
desvio_padrao = df_inativos['Idade'].std() #desvio padrão

print(f'Variancia: {variancia}')
print(f'Desvio Padrão: {desvio_padrao}')
```

```python
'''
Calculem as medidas de Dispersão da renda mensal
'''
variancia_renda = df_inativos['Renda mensal'].var() #variancia da renda
desvio_padrao_renda = df_inativos['Renda mensal'].std() #desvio padrão da renda

print(f'Variancia da Renda Mesal: {variancia_renda}')
print(f'Desvio Padrão da Renda Mesal: {desvio_padrao_renda}')
```

# Medidas Separatries - Inativos

```python
df_inativos.describe()
```

```python
# Quartis
q1 = df_inativos['Idade'].quantile(0.25)
q2 = df_inativos['Idade'].quantile(0.50)
q3 = df_inativos['Idade'].quantile(0.75)

# Min e Max
minimo = df_inativos['Idade'].min()
maximo = df_inativos['Idade'].max()

# Amplitude iterquartil
aiq = q3 - q1

#Limites Superoires e Inferiores
li = q1 - (1.5 * aiq)
ls = q3 + (1.5 * aiq)

print(f'Q1: {q1}')
print(f'Q2: {q2}')
print(f'Q3: {q3}')
print(f'Minimo: {minimo}')
print(f'Maximo: {maximo}')
print(f'Aiq: {aiq}')
print(f'Li: {li}')
print(f'Ls: {ls}')
```

```python
outlier_abaixo = df_inativos[df_inativos['Idade'] < li]
outlier_acima = df_inativos[df_inativos['Idade'] > ls]

display(outlier_abaixo)
display(outlier_acima)
```

```python
'''
calculem as medidas separatrizes da renda mensal
'''
# Quartis
q1 = df_inativos['Renda mensal'].quantile(0.25)
q2 = df_inativos['Renda mensal'].quantile(0.50)
q3 = df_inativos['Renda mensal'].quantile(0.75)

# Min e Max
minimo = df_inativos['Renda mensal'].min()
maximo = df_inativos['Renda mensal'].max()

# Amplitude iterquartil
aiq = q3 - q1

#Limites Superoires e Inferiores
li = q1 - (1.5 * aiq)
ls = q3 + (1.5 * aiq)

print(f'Q1: {q1}')
print(f'Q2: {q2}')
print(f'Q3: {q3}')
print(f'Minimo: {minimo}')
print(f'Maximo: {maximo}')
print(f'Aiq: {aiq}')
print(f'Li: {li}')
print(f'Ls: {ls}')
```

```python
outlier_acima = df_inativos[df_inativos['Renda mensal'] > ls]

display(outlier_acima)
```

```python
freq_rel = df_inativos['Cargo'].value_counts(normalize=True)
display(round(freq_rel*100,2).astype(str) + '%')
```

```python
freq_rel_formatada = round(freq_rel*100,2).astype(str) + '%'
freq = pd.concat([freq_abs, freq_rel_formatada], axis=1)
freq.columns = ['Frequência Absoluta','Frequência Relativa']
display(freq)
```

```python
limites = [0, 10, 20, 30, 40, 50]
labels = ['0-9','10-19','20-29','30-39','40-49']

df_inativos['Faixa de Anos na Empresa'] = pd.cut(
    df_inativos['Total de anos trabalhados na empresa'], # Fonte dos dados
    bins = limites, # São os nossos limites
    labels = labels # Rótulos exibidos na tela
)

freq_abs_num = df_inativos['Faixa de Anos na Empresa'].value_counts()
display(freq_abs_num)
```

```python
freq_rel_num = df_inativos['Faixa de Anos na Empresa'].value_counts(normalize=True)
display(round(freq_rel_num*100,2).astype(str) + '%')
```

```python
freq_anos = pd.DataFrame({
    'Absoluta': freq_abs_num,
    'Relativa': round(freq_rel_num*100,2).astype(str) + '%',
    'Frequencia Acumulada': freq_abs_num.cumsum(),
    'Frequencia Relativa Acumulada': round(freq_rel_num.cumsum()*100,2).astype(str) + '%'
})
display(freq_anos)
```

```python
'''
Grafico de barras:

Usamos para visualizar a frequência de categorias.
Normalmente, o nosso objetivo é entender quais categorias tem mais ou menos
registros.

Ex: Contar a quantidade de funcionários que pediram demissão
'''
# countplot - É um gráfico de barras que conta a quantidade de ocorrências
# x='Departamento' - Sinalizamos de onde vão ser retiradas as categorias
# data=df_inativos - De qual dataframe vamos tirar a informação
# order=df_inativos.value_counts() - Determina a ordem e realiza a contagem de cada valor
# .index - Retorna o índice da contagem, no nosso caso, o nome do departamento
sns.countplot(x='Departamento', data=df_inativos, order=df_inativos['Departamento'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Quantidade de Funcionários Inativos por Departamento')
plt.show()
```

```python
'''
Faça um grafico de barras retornando as informações sobre a Satisfação com o trabalho dos funcionários
'''
sns.countplot(x='Satisfação com o trabalho',
              data=df_inativos,
              order=df_inativos['Satisfação com o trabalho'].value_counts().index)

#plt.gca() - Pegamos o eixo atual
#containers[0] - Acessa o primeiro grupo de barras do gráfico
#fmt='%d' - formata para inteiro
plt.bar_label(plt.gca().containers[0], fmt='%d')
plt.title('Nivel de Satisfação com o trabalho de Inativos')
plt.show()
```

```python
'''
Faça um grafico de barras retornando as informações sobre a Satisfação com o trabalho dos funcionários
'''
sns.countplot(x='Satisfação com o trabalho',
              data=df_inativos,
              order=df_inativos['Satisfação com o trabalho'].value_counts().index)
total = len(df_inativos)
contagem = df_inativos['Satisfação com o trabalho'].value_counts()
#plt.gca() - Pegamos o eixo atual
#containers[0] - Acessa o primeiro grupo de barras do gráfico

plt.bar_label(plt.gca().containers[0],
              labels = [f'{(count/total * 100):.1f}%' for count in contagem])

plt.title('Distribuição de Funcionários Inativos por Satisfação com o Trabalho')
plt.show()
```

```python
'''
Vamos ver a distribuição por cargo
'''
sns.countplot(x='Cargo',
              data=df_inativos,
              order=df_inativos['Cargo'].value_counts().index,
              palette='viridis', # define a cor da coluna
              hue = 'Cargo') #Especifica qual a coluna será palicada
plt.xticks(rotation=45)
plt.title('Quantidade de Funcionários Inativos por Cargo')
plt.show()
```

```python
'''
Gráfico de pizza

Ideal para quando queremos entender a proporção de cada categoria.
'''
df_inativos['Gênero'].value_counts().plot(kind='pie',
                                          autopct = '%1.1f%%',
                                          colors=['#836FFF','#8FBC8F'],
                                          ylabel='')
plt.title('Distribuição de Funcionários Inativos por Gênero')
plt.show()
```

```python
'''
Descubram o percentual de funcionario que fazem viagem de negocio.

coluna = 'Viagem de Negócio'
autopct = '%1.1f%%' = Formatação do rotulo
colors=['#FF0000','#008000','#0000FF'] = Define as cores do gráfico
ylabel='' = Remove a legenda do eixo y
'''
df_inativos['Viagem de negócios'].value_counts().plot(kind='pie',
                                          autopct = '%1.1f%%',
                                          colors=['#FF0000','#008000','#0000FF'],
                                          ylabel='')
plt.title('% funcionarios Inativos que fazem viagem de negocio')
plt.show()
```
